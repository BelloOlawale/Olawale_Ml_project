```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_train = train_set.copy() # creating a copy of the train set for further eda
```

```python
# plotting histograms for all numerical features
import matplotlib.pyplot as plt
df_train.hist(bins=50, figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
# Working with the train set and only use the test set for validation.
print(f"The shpae of the train set is {train_set.shape} and the shape of the set is {test_set.shape}")
```

```python
# looking for correlations in the dataset
corr_matrix = df_train.corr(numeric_only=True)

# how each features correlate with the target
corr_matrix["previous"].sort_values(ascending=False)
```

```python
# Heapmap
# Generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = 1

# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(corr_matrix * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f',
            mask=mask)
plt.tight_layout()
plt.show()
```
